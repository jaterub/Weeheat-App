```python
df_csalud = pd.read_csv("../data/madrid/Centros atencion medica/centros_salud_filtrado.csv")
df_iglesias = pd.read_csv("../data/madrid/Iglesias/iglesias_filtradas.csv")
df_monumentos = pd.read_csv("../data/madrid/Monumentos/Monumentos_filtrados.csv")
df_museos = pd.read_csv("../data/madrid/museos/museos filtrados.csv")
df_salas = pd.read_csv("../data/madrid/Salas espectaculos artisticos/salas_espectaculos_filtradas.csv")
df_templo_no_cat = pd.read_csv("../data/madrid/Templos_no_catolicos/Templos_filtrados.csv")
df_fuentes = pd.read_csv("../data/madrid/fuentes/fuentes_nuevo_reduced.csv")
df_parques = pd.read_csv("../data/madrid/Parques y jardines/parques_y_jardines_nuevo_reduced.csv")
df_oficinas = pd.read_csv("../data/madrid/oficinas_turismo/tourist_office_nuevo_reduced.csv")
df_piscinas = pd.read_csv("../data/madrid/Piscinas/piscinas_madrid_nuevo_reduced.csv")
```

```python
df_concat_reduced = pd.concat([df_fuentes, df_museos, df_parques, df_oficinas, df_piscinas,df_templo_no_cat,df_salas,df_monumentos,df_iglesias,df_csalud], ignore_index=True)
df_concat_reduced
```

```python
df_concat_reduced['TIPO'] = df_concat_reduced['TIPO'].str.upper()
```

```python
df_concat_reduced["TIPO"].unique()
```

```python
df_concat_reduced.loc[df_concat_reduced['TIPO'] == 'SALAS ESPECTACULOS', 'TIPO'] = 'SALAS_ESPECTACULOS'
```

```python
df_concat_reduced["Unnamed: 0"].unique()
```

```python
df_concat_reduced.drop("Unnamed: 0",axis = 1, inplace= True)
```
