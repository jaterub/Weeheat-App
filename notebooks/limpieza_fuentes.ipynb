No hemos podido pasar los datos a Float porque hay un campo tanto en coordenada X como en y con el texto *no aplica*. Vamos a filtrar aquellos campos en los que se indique *no aplica*.

```python
df_fuentes[df_fuentes["COORD, X"] == 'no aplica']
```

Hemos preguntado a ChatGPT como podemos obtener los datos de *COORD, X* y *COORD, Y*, obteniendo la siguiente respuesta:

Utilizando las coordenadas geográficas proporcionadas (longitud: -3.693159, latitud: 40.414374) y el sistema de coordenadas proyectadas "ETRS89 / UTM Zona 30N", la conversión da como resultado las siguientes coordenadas proyectadas COORDX y COORDY:

*COORD, X*: 441131.329643776

*COORD, Y*: 4473173.342050204

Se ha comprobado con otros datos en el que sí tenemos *COORD, X*, *COORD, Y*, *latitud* y *longitud*, que el resultado que da ChatGPT es correcto (es decir, le hemos pasado únicamente los datos de latitud y longitud y nos ha devuelto el mismo resultado de *COORD, X* y *COORD, Y* que está presente en los datos).

```python
df_fuentes["COORD, X"] = df_fuentes["COORD, X"].str.replace("no aplica","441131.329643776")
df_fuentes["COORD, Y"] = df_fuentes["COORD, Y"].str.replace("no aplica","4473173.342050204")
```

Una vez aplicado el cambio, ya podemos pasar los datos a tipo float.

```python
df_fuentes["COORD, X"] = df_fuentes["COORD, X"].astype(float)
df_fuentes["COORD, Y"] = df_fuentes["COORD, Y"].astype(float)
df_fuentes
```

A continuación, eliminamos las columnas de *FECHA DATOS* y *SISTEMA DE COORDENADAS*. La primera porque la fecha no es interesante para nuestro proceso y la segunda porque solo existe un valor.

```python
def concatenar_fila(row):
    return f"{row['DISTRITO']}, {row['DIRECCION']}"

# Crear la nueva columna "DIRECCION" aplicando la función a lo largo de las filas
df_fuentes['DIRECCION'] = df_fuentes.apply(concatenar_fila, axis=1)
```

```python
nuevo_orden_columnas = ["TIPO","NOMBRE","DIRECCION","DESCRIPCION","LATITUD","LONGITUD"] 
df_fuentes = df_fuentes.reindex(columns=nuevo_orden_columnas)
```
