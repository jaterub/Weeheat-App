```python
import requests
import pandas as pd
```

```python
def obtener_piscinas():
    url = "https://datos.madrid.es/egob/catalogo/210227-0-piscinas-publicas.csv"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica si la solicitud fue exitosa

        # Si la solicitud es exitosa, retorna el contenido en formato CSV
        return response.content

    except requests.exceptions.RequestException as e:
        print("Error al obtener las piscinas:", e)
        return None
```

```python
contenido_csv = obtener_piscinas()

# Verificar si se obtuvo el contenido y guardarlo en un archivo CSV
if contenido_csv:
    with open("../data/madrid/Piscinas/piscinas_madrid.csv", "wb") as archivo_csv:
        archivo_csv.write(contenido_csv)
    print("Se ha guardado el archivo piscinas_madrid.csv")
```

### Nuevo enfoque, quieren todo en un .json, unificaremos columnas y los propios json en cada cosa. Solo 4 columnas tipo,nombre,direccion,coordenada-x,coordenada-y latitud,longitud y descripcion

```python
def concatenar_fila(row):
    return f"{row['CLASE-VIA']} {row['NOMBRE-VIA']} {row['NUM']}, {row['LOCALIDAD']}, {row['CODIGO-POSTAL']}"

# Crear la nueva columna "DIRECCION" aplicando la función a lo largo de las filas
df['DIRECCION'] = df.apply(concatenar_fila, axis=1)
```

```python
df.drop(['PK','TRANSPORTE','CLASE-VIA', 'NOMBRE-VIA', 'NUM',
       'LOCALIDAD', 'CODIGO-POSTAL', 'BARRIO', 'DISTRITO'],axis = 1,inplace=True)
```
