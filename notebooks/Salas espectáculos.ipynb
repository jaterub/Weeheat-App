```python
import requests
import pandas as pd
```

```python
def obtener_teatros():
    url = "https://datos.madrid.es/egob/catalogo/208862-7650046-ocio_salas.csv"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica si la solicitud fue exitosa

        # Si la solicitud es exitosa, retorna el contenido en formato CSV
        return response.content

    except requests.exceptions.RequestException as e:
        print("Error al obtener los teatros:", e)
        return None
```

```python
contenido_csv = obtener_teatros()

# Verificar si se obtuvo el contenido y guardarlo en un archivo CSV
if contenido_csv:
    with open("../data/madrid/Salas espectaculos artisticos/teatros.csv", "wb") as archivo_csv:
        archivo_csv.write(contenido_csv)
    print("Se ha guardado el archivo teatros.csv")
```

```python
df_teatros = pd.read_csv("../data/madrid/Salas espectaculos artisticos/teatros.csv", sep=";", encoding="latin-1")
```
