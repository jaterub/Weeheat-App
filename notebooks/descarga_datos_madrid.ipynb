```python
import requests

# URL de la API de AEMET
url = "https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/2013-01-01T00%3A00%3A00UTC/fechafin/2017-01-01T00%3A00%3A00UTC/todasestaciones"

# Realizar la solicitud GET a la API
response = requests.get(url)

# Verificar si la solicitud fue exitosa (código de respuesta 200)
if response.status_code == 200:
    # Obtener la URL de los datos climatológicos desde la respuesta JSON
    data_url = response.json()["datos"]
    
    # Realizar una nueva solicitud para obtener los datos climatológicos
    data_response = requests.get(data_url)
    
    # Verificar si la solicitud fue exitosa (código de respuesta 200)
    if data_response.status_code == 200:
        # Obtener los datos climatológicos en formato JSON
        data = data_response.json()
        
        # Procesar los datos como desees, por ejemplo, imprimirlos
        print(data)
    else:
        print("Error al obtener los datos climatológicos:", data_response.status_code)
else:
    print("Error al obtener la URL de los datos climatológicos:", response.status_code)
```

```python
tmin_mean = df['tmin'].mean()
df['tmin'].fillna(tmin_mean, inplace=True)
tmin_mean = df['tmax'].mean()
df['tmax'].fillna(tmin_mean, inplace=True)
tmin_mean = df['tmed'].mean()
df['tmed'].fillna(tmin_mean, inplace=True)
```

Conversión data legible

```python
df['fecha'] = pd.to_datetime(df['fecha'])
df['tmed'] = (df['tmed']).str.replace(',','.')
df['tmin'] = (df['tmin']).str.replace(',','.')
df['tmax'] = (df['tmax']).str.replace(',','.')
df['tmed'] = (df['tmed']).astype(float)
df['tmin'] = (df['tmin']).astype(float)
df['tmax'] = (df['tmax']).astype(float)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['tmed'], label='Temperatura Media')
plt.plot(df['fecha'], df['tmin'], label='Temperatura Mínima')
plt.plot(df['fecha'], df['tmax'], label='Temperatura Máxima')

plt.xlabel('Fecha')
plt.ylabel('Temperatura (°C)')
plt.title('Timeline de Temperaturas')
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_monthly = df.groupby(df['fecha'].dt.to_period('M')).agg({
    'tmed': 'mean',
    'tmin': 'min',
    'tmax': 'max'
}).reset_index()

# Convertir las fechas de vuelta al formato de fecha para el gráfico
df_monthly['fecha'] = df_monthly['fecha'].dt.to_timestamp()

# Crear el gráfico de línea por meses
plt.figure(figsize=(10, 6))
plt.plot(df_monthly['fecha'], df_monthly['tmed'], label='Temperatura Media')
plt.plot(df_monthly['fecha'], df_monthly['tmin'], label='Temperatura Mínima')
plt.plot(df_monthly['fecha'], df_monthly['tmax'], label='Temperatura Máxima')

plt.xlabel('Fecha (Mes)')
plt.ylabel('Temperatura (°C)')
plt.title('Timeline de Temperaturas por Mes')
plt.legend()

plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad
plt.tight_layout()  # Ajustar el diseño del gráfico

plt.show()
```

```python
def convertir_a_entero(cadena):
    # Eliminar los espacios y dividir los valores múltiples
    valores = cadena.split(',')
    # Convertir cada valor en entero y devolverlos como una lista
    return [int(valor) for valor in valores]

# Aplicar la función a la columna 'mi_columna' y obtener una nueva columna 'columna_enteros'
df_museos['ACCESIBILIDAD'] = df_museos['ACCESIBILIDAD'].apply(convertir_a_entero)

# Imprimir el DataFrame resultante

```

Función para ver tiempo al momento

```python
import requests

API_KEY = "2a4888d5bb0e8e262434c0d42226a16e"
city = "Madrid"

url = f"https://api.openweathermap.org/data/2.5/weather?q=Madrid,es&appid={API_KEY}"

response = requests.get(url)

if response.status_code == 200:
    # La solicitud se realizó correctamente
    weather_data = response.json()
    temperature_kelvin = weather_data["main"]["temp"]
    temperature_celsius = temperature_kelvin - 273.15  # Convertir a Celsius
    print(f"La temperatura actual en {city} es {temperature_celsius:.2f} grados Celsius")
else:
    # La solicitud no se realizó correctamente
    print("Error al obtener los datos meteorológicos.")
```
