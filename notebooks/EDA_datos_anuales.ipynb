```python
columnas_dias = [f'D{i:02}' for i in range(1, 32)]

# Derretir las columnas de días y renombrar las columnas resultantes
data_melted = pd.melt(df_temperatura_acustica, id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], value_vars=columnas_dias, var_name='DIA', value_name='TEMPERATURA')

# Función personalizada para formatear la fecha correctamente
def format_fecha(row):
    try:
        fecha_str = f"{row['ANO']}-{row['MES']:02d}-{int(row['DIA'][1:]):02d}"
        return pd.to_datetime(fecha_str, format='%Y-%m-%d')
    except ValueError:
        return pd.NaT

# Aplicar la función para crear la columna 'Fecha'
data_melted['Fecha'] = data_melted.apply(format_fecha, axis=1)

# Eliminar las columnas innecesarias
data_melted = data_melted.drop(columns=['ANO', 'MES', 'DIA'])

# Ordenar el DataFrame por la columna 'Fecha'
data_melted = data_melted.sort_values(by='Fecha')

# Mostrar el resultado
print(data_melted)
```
