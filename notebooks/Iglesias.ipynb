```python
import requests
import pandas as pd
```

```python
def obtener_iglesias():
    url = "https://datos.madrid.es/egob/catalogo/209426-0-templos-catolicas.csv"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica si la solicitud fue exitosa

        # Si la solicitud es exitosa, retorna el contenido en formato CSV
        return response.content

    except requests.exceptions.RequestException as e:
        print("Error al obtener las iglesias:", e)
        return None
```

```python
contenido_csv = obtener_iglesias()

# Verificar si se obtuvo el contenido y guardarlo en un archivo CSV
if contenido_csv:
    with open("../data/madrid/Iglesias/iglesias_madrid.csv", "wb") as archivo_csv:
        archivo_csv.write(contenido_csv)
    print("Se ha guardado el archivo iglesias_madrid.csv")
```

```python
df['TIPO'] = df['TIPO'].str.extract(r'\/([^\/]+)$')
```

```python
df = df.drop(["DESCRIPCION-ENTIDAD","HORARIO","EQUIPAMIENTO","DESCRIPCION","ACCESIBILIDAD","CONTENT-URL","TIPO-NUM",
              "PLANTA","PUERTA","ESCALERAS","ORIENTACION","LOCALIDAD","PROVINCIA","CODIGO-POSTAL","COD-DISTRITO","COD-BARRIO",
              "TELEFONO","FAX","EMAIL","Unnamed: 32","PK","COORDENADA-X","COORDENADA-Y","BARRIO"],axis=1)
```

```python
df["NUM"] = df["NUM"].fillna("N/A o N/D")
```

```python
# Intercambiamos el orden de las columnas 3 y 4 para luego cambiarles el nombre para favorecer la visualización.
df.iloc[:, [2, 3]] = df.iloc[:, [3, 2]].values
```

```python
df.rename(columns={"NOMBRE-VIA": "CLASE-VIA", "CLASE-VIAL": "NOMBRE-VIA"}, inplace=True)
```

```python
df["NUM"] = df["NUM"].astype(str)
```

```python
df["DIRECCION"] = df["CLASE-VIA"] + " " + df["NOMBRE-VIA"] + " " + df["NUM"]
```

```python
df.drop(["CLASE-VIA","NOMBRE-VIA","NUM"],axis=1,inplace=True)
```

```python
df = df.loc[df["DISTRITO"].isin(["CENTRO", "CHAMBERI", "RETIRO","ARGANZUELA", "SALAMANCA"])]
```

```python
df["DESCRIPCION"] = df["TIPO"]
```

```python
df["TIPO"] = "Templos"
```

```python
df.drop(["TRANSPORTE","DISTRITO"],axis=1,inplace=True)
```

```python
df = df.reindex(columns=['TIPO', 'NOMBRE', 'DIRECCION', 'DESCRIPCION', 'LATITUD', 'LONGITUD'])
```
