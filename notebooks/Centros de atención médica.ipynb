```python
import requests
import pandas as pd
import re
```

```python
def obtener_centros_atencion_medica():
    url = "https://datos.madrid.es/egob/catalogo/212769-0-atencion-medica.csv"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica si la solicitud fue exitosa

        # Si la solicitud es exitosa, retorna el contenido en formato CSV
        return response.content

    except requests.exceptions.RequestException as e:
        print("Error al obtener los centros de atención médica:", e)
        return None
```

```python
contenido_csv = obtener_centros_atencion_medica()

# Verificar si se obtuvo el contenido y guardarlo en un archivo CSV
if contenido_csv:
    with open("../data/madrid/Centros atencion medica/centros_atencion_medica.csv", "wb") as archivo_csv:
        archivo_csv.write(contenido_csv)
    print("Se ha guardado el archivo centros_atencion_medica.csv")
```

```python
df = df.drop(["PK","DESCRIPCION-ENTIDAD","HORARIO","EQUIPAMIENTO","DESCRIPCION","ACCESIBILIDAD","CONTENT-URL",
              "TIPO-NUM","PLANTA","PUERTA","ESCALERAS","ORIENTACION","LOCALIDAD","PROVINCIA",
              "CODIGO-POSTAL","COD-BARRIO","COD-DISTRITO","TELEFONO","FAX","EMAIL","Unnamed: 32","COORDENADA-X","COORDENADA-Y"],axis=1)
```

```python
df.info()
```

```python
df['TIPO'] = df['TIPO'].str.extract(r'\/([^\/]+)$')
```

```python
df["TIPO"].unique()
```

```python
df["TRANSPORTE"] = df["TRANSPORTE"].fillna("No disponible")
df["NUM"] = df["NUM"].fillna("N/A o N/D")
df["BARRIO"] = df["BARRIO"].fillna("No disponible")
df["TIPO"] = df["TIPO"].fillna("No disponible")
```

```python
# Intercambiamos el orden de las columnas 3 y 4 para luego cambiarles el nombre para favorecer la visualización.
df.iloc[:, [2, 3]] = df.iloc[:, [3, 2]].values
```

```python
df.rename(columns={"NOMBRE-VIA": "CLASE-VIA", "CLASE-VIAL": "NOMBRE-VIA"}, inplace=True)
```

```python
df.info()
```

```python
df["NUM"] = df["NUM"].astype(str)
```

```python
df["DIRECCION"] = df["CLASE-VIA"] + " " + df["NOMBRE-VIA"] + " " + df["NUM"]
```

```python
df.drop(["CLASE-VIA","NOMBRE-VIA","NUM"],axis=1,inplace=True)
```

```python
df = df.loc[df["DISTRITO"].isin(["CENTRO", "CHAMBERI", "RETIRO","ARGANZUELA", "SALAMANCA"])]
df = df.loc[df["TIPO"].isin(["CentrosSalud", "OtrosCentrosMedicos", "HospitalesClinicasSanatorios"])]
```

```python
df["TIPO"] = "Salud"
```

```python
df.drop(["TRANSPORTE","BARRIO","DISTRITO"],axis=1,inplace=True)
```

```python
df = df.reindex(columns=['TIPO', 'NOMBRE', 'DIRECCION', 'DESCRIPCION', 'LATITUD', 'LONGITUD'])
```
