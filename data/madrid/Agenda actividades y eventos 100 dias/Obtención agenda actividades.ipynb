```python
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
fecha_actual = datetime.now()
fecha_actual
```

```python
fecha_actual_str = fecha_actual.strftime("%Y-%m-%d")
fecha_actual_str
```

```python
def obtener_agenda_activades_eventos_100dias():
    url = "https://datos.madrid.es/egob/catalogo/300107-0-agenda-actividades-eventos.csv"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica si la solicitud fue exitosa

        # Si la solicitud es exitosa, retorna el contenido en formato CSV
        return response.content

    except requests.exceptions.RequestException as e:
        print("Error al obtener los eventos culturales:", e)
        return None
```

```python
contenido_csv = obtener_agenda_activades_eventos_100dias()

# Verificar si se obtuvo el contenido y guardarlo en un archivo CSV
if contenido_csv:
    nombre_archivo = f'Agenda activades 100 dias a {fecha_actual_str}.csv'

    # Guardar el contenido en el archivo CSV
    with open(nombre_archivo, "wb") as archivo_csv:
        archivo_csv.write(contenido_csv)
    print(nombre_archivo)
```

```python
df = pd.read_csv(nombre_archivo, sep=";", encoding="latin-1")
```

```python
# Calcular la fecha de hoy más 10 días
fecha_maxima = fecha_actual + timedelta(days=10)
fecha_maxima
```
